```python
import jieba
```

```python
text = "小明毕业于北京大学计算机系"
```

```python
# 1. 默认精确模式
words_generator = jieba.cut(text)  # 返回一个生成器，节省内存
for word in words_generator:
    print(word)
word_list = jieba.lcut(text) # 返回一个列表
print(word_list)
```

```python
# 2.全模式
jieba.lcut(text,cut_all = True)
```

```python
# 3.搜索引擎模式
jieba.lcut_for_search(text)
```

```python
# 4.自定义词典
# 自定义词典以便包含 jieba 词库里没有的词，用于增强特定领域词汇的识别能力。
# 1.加载词典
jieba.load_userdict('./jieba_self_dict.txt')
jieba.lcut("随着云计算技术的普及，越来越多企业开始采用云原生架构来部署服务，并借助大模型能力提升智能化水平，实现业务流程的自动化与智能决策。")
```
